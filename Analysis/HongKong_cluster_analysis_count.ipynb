# Here shows the count method

First, Import necessary library

```python
import numpy as np
import pandas as pd
import io
import folium # map rendering library
import matplotlib.cm as cm
import matplotlib.colors as colors
```

Next, import the dataset and combine it into one

```python
lists_0=pd.read_csv(r'grouped_lists_0.csv')
lists_1=pd.read_csv(r'grouped_lists_1.csv')
lists_2=pd.read_csv(r'grouped_lists_2.csv')
lists_3=pd.read_csv(r'grouped_lists_3.csv')
lists_4=pd.read_csv(r'grouped_lists_4.csv')
lists_1
```

```python
rawdata = pd.concat([lists_0,lists_1,lists_2,lists_3,lists_4])
rawdata
```

In the followings, perform filtering to remove location without well defined coordinate, either is NaN value or the coordinate is outsite of Hong Kong region.

```python
rawdata = rawdata.dropna()
#HK lands departments
#Hong Kong Special Administrative Region lies between Latitude 22Â°08' North and 22Â°35' North, Longitude 113Â°49' East and 114Â°31' East
#remove coordinates not lies in Hong Kong 22<latitute<23, 113<longitude<115
rawdata = rawdata[rawdata.lat_col >22]
rawdata = rawdata[rawdata.lat_col <23]
rawdata = rawdata[rawdata.long_col >113]
rawdata = rawdata[rawdata.long_col <115]
rawdata
```

Sort the dataframe based on count value, then list out the top 10 building having the most count value.

```python
rawdata_sort= rawdata.sort_values(by=['count'],ascending=False)
rawdata_sort = rawdata_sort[0:10]
rawdata_sort
```

Plot the data in map. The Hong Kong base map is used and each point is drawn in blue circle where the radius is the count value divided by 2. This map shows the top 10 buildings where cases mostly occurs. 
a html file count.html is saved for further view on the cases.

```python
latitude =22.329
longitude = 114.169
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to the map
for lat, lon, poi,count in zip(rawdata_sort['lat_col'], rawdata_sort['long_col'], rawdata_sort['Building name'],rawdata_sort['count']):
    label = folium.Popup(str(poi), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=count/2,
        popup=label,
        fill=True,
        fill_opacity=0.7).add_to(map_clusters)
    
map_clusters.save('count.html')
map_clusters
```
